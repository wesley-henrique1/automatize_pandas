```python
from modulos._settings import Relatorios, Outros, Output
import datetime as dt
import pandas as pd
import numpy as np
import warnings
import os
import re
warnings.simplefilter(action='ignore', category=UserWarning)

list_path= [Relatorios.rel_96, Outros.ou_end]
```

```python
colunas_origem = [
    "RUA"
    ,"PREDIO"
    ,"CODPROD"
    ,"OBS2"
    ,"ABASTECEPALETE"
    ,"CAPACIDADE"
    ,"PONTOREPOSICAO"
    ,"QTUNITCX"
    ,"LARGURAARM"
    ,"COMPRIMENTOARM"
    ,"LASTROPAL"
    ,"PK_END"
    ,"DESCRICAO"
    ,"QTTOTPAL"
    ,"CARACTERISTICA"
    ,"APTO"
    ,"TIPO_1"
]
dados_prod = pd.read_excel(list_path[0], usecols= colunas_origem)
endereco = pd.read_excel(list_path[1], sheet_name= 'STATUS', usecols= ["RUA", "TIPO_RUA", "CARACT"])
TEMP = dados_prod.merge(endereco, on= 'RUA', how= 'inner')
df_prod = TEMP.loc[TEMP['RUA'].between(1,39)].copy()
dic_raname = {
    'ABASTECEPALETE' : 'FLEG_ABST',
    'CAPACIDADE' : 'CAP',
    'PONTOREPOSICAO' : 'P_REP',
    'QTUNITCX' : 'FATOR'
}

list_meio = ['3-MEDIO (0,80)', '7-MEIO PALETE']
```

```python
df_PRODUTO = df_prod.rename(columns=dic_raname)
df_PRODUTO['TESTE'] = df_PRODUTO['RUA'].astype(str) + " - " + df_PRODUTO['PREDIO'].astype(str)

filtro = (df_PRODUTO['APTO'].between(100, 199)) & (df_PRODUTO['PK_END'].isin(list_meio))
df_PRODUTO.loc[filtro, 'qt_meio'] = df_PRODUTO[filtro].groupby('TESTE')['TESTE'].transform('count')
```
