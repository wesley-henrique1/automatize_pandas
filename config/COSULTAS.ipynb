```python
def ajuste_col(df_original, col):
            df_copia = df_original.copy() 
            for i in col:
                df_copia[i] = df_copia[i].fillna(0).astype(str)
                df_copia[i] = df_copia[i].str.replace(".", "", regex=False)
                df_copia[i] = df_copia[i].str.replace(",", ".")
                df_copia[i] = df_copia[i].fillna(0).astype(float)
            return df_copia
```

```python
caso = df_1707[['COD','DESCRICAO','RUA', 'PREDIO', 'NIVEL', 'APTO','PICKING', 'PULMAO','ENDERECO', 'GERENCIAL']].copy()
caso = ajuste_col(caso, ['ENDERECO', 'GERENCIAL'])
caso[['RUA', 'PREDIO', 'NIVEL', 'APTO','ENDERECO', 'GERENCIAL']] = caso[['RUA', 'PREDIO', 'NIVEL', 'APTO','ENDERECO', 'GERENCIAL']].fillna(0).astype(int)
caso['DIFERENCA'] = caso['ENDERECO'] - caso['GERENCIAL']
```
